```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'age':[23,18,90,25,87,54,89]
})
```

```python
df['Age_bin_width'] = pd.cut(df['age'], bins=4)
```

```python
df
```

```python
df['Age_bin_width'] = pd.qcut(df['age'], q=6)
```

```python
df
```

```python
#custom binding
bins = [0, 17, 45, 100]
labels = ['Youth','Adult','Senior']
df['Age_bin_custom'] = pd.cut(df['age'], bins=bins, labels=labels)

print(df)
```
