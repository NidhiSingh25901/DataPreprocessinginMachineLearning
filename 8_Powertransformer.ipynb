```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PowerTransformer
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    'skewed': np.random.exponential(scale=2, size=1000)
})
```

```python
data['skewed'].hist(bins=30)
plt.title("Before Power Transformation")
plt.show()
```

```python
pt = PowerTransformer(method='yeo-johnson')
data['transformed'] = pt.fit_transform(data['skewed'].values.reshape(-1,1))
```

```python
data['transformed'].hist(bins=30)
plt.title("After Power Transformation")
plt.show()
```
